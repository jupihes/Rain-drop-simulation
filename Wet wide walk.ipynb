```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

## constants
sidewalk_len = 10000
drop_len = 100

## drop generation
def drop():
	temp = 0 
	temp = np.random.randint(0,sidewalk_len-drop_len,size=1)[0]
	#print([temp,temp+100])
	return [temp,temp+100,(temp+100- temp)]
```

```python
def wet_length_cal(drop_list):
	drop_list.sort()
	wet_sum = 0 
	for i in range(1,len(drop_list)):
	
		#print(drop_list[i])
		if drop_list[i-1][1] < drop_list[i][0] :
			wet_sum = wet_sum + drop_list[i][1] -drop_list[i][0]
		if drop_list[i-1][1] >= drop_list[i][0] :
			wet_sum = wet_sum + drop_list[i][1] -drop_list[i-1][1]


	#print(len(drop_list),wet_sum,wet_sum/sidewalk_len)
	return [len(drop_list),wet_sum,wet_sum/sidewalk_len]
```
